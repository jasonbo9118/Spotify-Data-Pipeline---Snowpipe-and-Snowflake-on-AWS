```python
from dotenv import load_dotenv
import os
import spotipy
from spotipy.oauth2 import SpotifyOAuth

load_dotenv()

client_id = os.getenv("SPOTIPY_CLIENT_ID")
client_secret = os.getenv("SPOTIPY_CLIENT_SECRET")
redirect_uri = os.getenv("REDIRECT_URI")

sp = spotipy.Spotify(
    auth_manager=SpotifyOAuth(
        client_id=client_id,
        client_secret=client_secret,
        redirect_uri=redirect_uri,
        scope="playlist-read-private"
    )
)

playlist_link = "https://open.spotify.com/playlist/3k4RaZjxSzRIRlrtUEaGka"

playlist_id = playlist_link.split("/")[-1]

spotify_data = sp.playlist_items(
    playlist_id
)
```

```python
album_list = []
for row in spotify_data['items']:
    album_id = row['item']['album']['id']
    album_name = row['item']['album']['name']
    album_release_date = row['item']['album']['release_date']
    album_total_tracks = row['item']['album']['total_tracks']
    album_uri = row['item']['album']['uri']
    album_element = {'album_id' : album_id, 'album_name' : album_name, 'album_release_date' : album_release_date, 'album_total_tracks' : album_total_tracks, 'album_uri' : album_uri}
    album_list.append(album_element)
```

```python
album_list
```

```python
artist_list = []
for row in spotify_data['items']:
    artist_id = row['item']['artists'][0]['id']
    artist_name = row['item']['artists'][0]['name']
    artist_uri = row['item']['artists'][0]['uri']
    artist_element = {'artist_id' : artist_id, 'artist_name' : artist_name, 'artist_uri' : artist_uri}
    artist_list.append(artist_element)
```

```python
song_list = []
for row in spotify_data['items']:
    song_id = row['item']['id']
    song_name = row['item']['name']
    song_duration_ms = row['item']['duration_ms']
    song_explicit = row['item']['explicit']
    album_id = row['item']['album']['id']
    artist_id = row['item']['artists'][0]['id']
    song_elements = {'song_id' : song_id, 'song_name' : song_name, 'song_duration_ms' : song_duration_ms, 'song_explicit' : song_explicit, 'album_id' : album_id, 'artist_id' : artist_id}
    song_list.append(song_elements)
```
